**DUPLICATE FINDINGS**

```python

# Duplicate Records

# Countig the duplicates
duplicate_count = retail_df.duplicated().sum()
print("Total Duplicate Rows:", duplicate_count)

# Display sample duplicates
if duplicate_count > 0:
    display(retail_df[retail_df.duplicated()].head())

# Removing duplicates values
retail_df = retail_df.drop_duplicates()
print("\nDuplicates removed. New shape:", retail_df.shape)
```

Detect Outliers

```python
num_cols = retail_df.select_dtypes(include=['int64', 'float64']).columns

total_outliers = 0  # To count total outliers in dataset
outlier_details = {}  # To store which column has outliers

for col in num_cols:
    Q1 = retail_df[col].quantile(0.25)
    Q3 = retail_df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = retail_df[(retail_df[col] < lower) | (retail_df[col] > upper)]
    count = len(outliers)

    if count > 0:
        outlier_details[col] = count
        total_outliers += count

# ===========================
#       FINAL OUTPUT
# ===========================

if total_outliers > 0:
    print(f"\nTotal Outliers in Entire Dataset: {total_outliers}\n")
    print("Columns with Outliers:")
    for col, cnt in outlier_details.items():
        print(f"- {col}: {cnt} outliers")
else:
    print(" No outliers found in the entire dataset!")
```

## Univariate Analysis - Numerical Features


```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols_retail = ['Quantity_numeric', 'Price', 'TotalPrice', 'StockCode_length', 'StockCode_word_count']

for col in numerical_cols_retail:
    print(f"\n--- Descriptive Statistics for {col} ---")
    display(retail_df[col].describe())

    plt.figure(figsize=(14, 6))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(retail_df[col], kde=True, bins=50)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

    # Box plot
    plt.subplot(1, 2, 2)
    sns.boxplot(y=retail_df[col])
    plt.title(f'Box Plot of {col}')
    plt.ylabel(col)

    plt.tight_layout()
    plt.show()
```

## Univariate Analysis - Categorical Features

### Subtask:

```python
categorical_cols_retail = ['Country', 'Description']

for col in categorical_cols_retail:
    print(f"\n--- Value Counts for {col} ---")
    if col == 'Description':
        # Display top 30 descriptions for readability due to high cardinality
        value_counts = retail_df[col].value_counts().head(30)
        display(value_counts)

        plt.figure(figsize=(12, 7))
        sns.barplot(x=value_counts.index, y=value_counts.values, palette='viridis')
        plt.title(f'Top 30 Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.xticks(rotation=90, ha='right')
        plt.tight_layout()
        plt.show()

    else:
        value_counts = retail_df[col].value_counts()
        display(value_counts)

        plt.figure(figsize=(12, 7))
        sns.barplot(x=value_counts.index, y=value_counts.values, palette='viridis')
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.xticks(rotation=90, ha='right')
        plt.tight_layout()
        plt.show()
```
